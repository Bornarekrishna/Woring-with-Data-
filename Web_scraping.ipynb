```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
wp = requests.get('https://www.flipkart.com/search?q=samsung+mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_9_sc_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_9_sc_na_ps&as-pos=1&as-type=RECENT&suggestionId=samsung+mobiles%7CMobiles&requestId=255ec6c6-8331-4ca1-8c67-652956cbc66f&as-backfill=on',headers=headers).text
```

```python
soup=BeautifulSoup(wp,'lxml')
```

```python
mobiles=soup.find_all('div',class_='_2kHMtA')
```

```python
len(mobiles)
```

```python
name=[]
rating=[]
ori_price=[]
price=[]
highlight=[]

for i in mobiles:

  name.append(i.find('div',class_='_4rR01T').text.strip())
  rating.append(i.find('div',class_='_3LWZlK').text.strip())
  ori_price.append(i.find('div',class_='_3I9_wc _27UcVY').text.strip())
  price.append(i.find('div',class_='_30jeq3 _1_WHN1').text.strip())
  highlight.append(i.find('ul').text.strip())
```

```python
df=pd.DataFrame({'Mobile':name,'Rating':rating,'Ori_price':ori_price,'Price':price,'Highlight':highlight})
```
